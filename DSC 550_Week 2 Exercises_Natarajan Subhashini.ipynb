```python
# Load required libraries
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Change directory
os.chdir("/Users/manikantanvenkataraman/Downloads")
os.getcwd()
```

```python
# Read food group data into a dataframe

foodgrp_df = pd.read_excel("/Users/manikantanvenkataraman/Downloads/US Food Import.xlsx")
foodgrp_df.head()
```

```python
# Bar plot 

foodgrp_df_sorted=foodgrp_df.sort_values(2021) # Sort the dataframe based on values in 2021
x=foodgrp_df_sorted.Food_Group #Assign Food group to x
y=foodgrp_df_sorted[2021] # assign 2021 values to y


plt.barh(x,y) # Plot a horizontal bar chart 
plt.xlabel("Import Value in million USD") # Assign xlabel
plt.ylabel("Imported food group") #Assign ylabel
plt.title("Value of food group imported in 2021") #Assign title
plt.show() # Show the labels
```

```python
# Line plot
foodgrp_df1=foodgrp_df.drop(columns=['Unit']) # Create a new df dropping the field unit
foodgrp_df1.set_index('Food_Group',inplace=True) # set Food Group as the index
foodgrp_df_tr=foodgrp_df1.transpose() # Transpose the rows and columns 
# Rename the column names
foodgrp_df_tr.columns=['Live_Meat_Animals','Meats', 'Fish_Shellfish', 'Dairy','Vegetables', 'Fruits', 'Nuts', 'Coffee_tea_spices', 'Grains', 'Vegetable_oils', 'Sugar_Candy','Cocoa_Chocolate', 'Other', 'Beverages'] 

# Plot the line graph
plt.plot(foodgrp_df_tr.Fish_Shellfish, label='Fish_Shellfish')
plt.plot(foodgrp_df_tr.Vegetables, label='Vegetables')
plt.plot(foodgrp_df_tr.Fruits, label='Fruits')
plt.plot(foodgrp_df_tr.Grains, label='Grains')
plt.plot(foodgrp_df_tr.Other,label='Other edible products')
plt.plot(foodgrp_df_tr.Beverages, label='Beverages')
plt.xlabel('Year') # Assign xlabel
plt.ylabel('Import value in Million$') # Assign ylabel
plt.title('Import Value Trend') # Assign title
plt.legend() # Add legends
```

```python
# Pie chart

y=foodgrp_df[2021] # Assign 2021 values to y
mylabels=foodgrp_df.Food_Group # Assign the foof group as mylabels
# Plot the pie chart with labels and percentages
plt.pie(y, labels=mylabels,autopct='%1.0f%%', pctdistance=0.8, labeldistance=1.2)
plt.show()
```
